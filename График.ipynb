```python
# Random test data
np.random.seed(1699)

data0 = [[3.5, 3.7],[8.9,9,9,9,10,9.8],[0.5,0.6], [1.2,1.3,1.3]]
data1 = [[3.5, 3.7],[7.7,8.1,7.9,8,8.3],[0.5,0.6], [1.2,1.3]]
data2 = [[0.75, 0.7,0.74],[2.7,3.1,2.9,3],[-0.1,-0.05, -0.07, -0.08, -0.12], [0.1,0.15, 0.07, 0.05, 0.2]]
labels = ["CHN", "IND", "USA", "E_U"]

fig, ax = plt.subplots(nrows=1, ncols=3,  figsize=(15, 5), tight_layout=True)

# от 0 до 10, шаг 1
bplot1 = ax[0].boxplot(data0, vert=True,  patch_artist=True, labels=labels)

ax[0].set_xlabel('Анаконды')
ax[0].set_ylabel('Очень длинная подпись, очень непонятная')
ax[0].text(0.6, 9.7, r'a', fontsize=25, fontweight='bold')
ax[0].yaxis.set_minor_locator(ticker.MultipleLocator(0.25))

bplot2 = ax[1].boxplot(data1, vert=True, patch_artist=True, labels=labels) 

ax[1].set_xlabel('Пингвины')
ax[1].text(0.6, 8, r'b', fontsize=25, fontweight='bold')
ax[1].yaxis.set_minor_locator(ticker.MultipleLocator(0.25))

bplot3 = ax[2].boxplot(data2, vert=True, patch_artist=True, labels=labels)

ax[2].set_xlabel('Яшперицы')
ax[2].text(0.6,3, r'c', fontsize=25, fontweight='bold')
ax[2].hlines(0, 0, 4, linewidth = 1)
ax[2].yaxis.set_minor_locator(ticker.MultipleLocator(0.1))

plt.show()
```
